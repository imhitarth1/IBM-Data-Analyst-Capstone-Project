### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
country = pd.DataFrame(df['Country'].value_counts().reset_index())
country.to_csv('country.csv', index = False)
```

```python
country.loc[180, 'Country']
```

```python
country_mapping = {
    'U.S.A.' : ['United States of America'],
    'U.K.' : ['United Kingdom of Great Britain and Northern Ireland'],
    'Russia' : ['Russian Federation'],
    'Iran': ['Iran, Islamic Republic of...'],
    'Hong Kong' : ['Hong Kong (S.A.R.)'],
    'U.A.E.' : ['United Arab Emirates'],
    'Venezuela' : ['Venezuela, Bolivarian Republic of...'],
    'South Korea' : ['Republic of Korea'],
    'Moldova' : ['Republic of Moldova'],
    'North Macedonia' : ['Republic of North Macedonia'],
    'Syria' : ['Syrian Arab Republic'],
    'Tanzania' : ['United Republic of Tanzania'],
    'Republic of Congo' : ['Congo, Republic of the...'],
    'North Korea' : ['Democratic People\'s Republic of Korea', 'North Korea'],
    'Micronesia' : ['Micronesia, Federated States of...']
}
```

```python
def standardize_country(entry):
    if pd.isnull(entry):
        return None
    entry_lower = entry.lower()
    for standard, variants in country_mapping.items():
        if any(variant.lower() in entry_lower for variant in variants):
            return standard
    return entry 
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
df['Country'] = df['Country'].apply(standardize_country)
```

```python
most_freq_val = df['Country'].mode()
df['Country'] = df['Country'].fillna(most_freq_val[0])
```

```python
## Write your code here
most_freq_val_rw = df['RemoteWork'].mode()
df['RemoteWork'] = df['RemoteWork'].fillna(most_freq_val_rw[0])
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
df['ConvertedCompYearly_Norm'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['Log_ConvertedCompYearly'] = np.log(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
df['YearsCodePro'].isnull().sum()
```

```python
df['YearsCodePro'].unique()
```

```python
## Write your code here
df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year' : '<1', 'More than 50 years' : '50+'})
df['YearsCodePro'].value_counts()
```

```python
def map_senior(years):
    if pd.isna(years):
        return np.nan
    elif years == '<1':
        return 'Beginner'
    elif years in ['1', '2']:
        return 'Junior'
    elif years in ['3', '4', '5']:
        return 'Intermediate'
    elif years in [str(i) for i in range(6, 11)]:
        return 'Senior'
    elif years in [str(i) for i in range(11, 21)]:
        return 'Lead'
    elif years in [str(i) for i in range(21, 31)]:
        return 'Principal'
    else:
        return 'Veteran'
```

```python
df['SeniorLevel'] = df['YearsCodePro'].apply(map_senior)
df['SeniorLevel'].value_counts()
```
